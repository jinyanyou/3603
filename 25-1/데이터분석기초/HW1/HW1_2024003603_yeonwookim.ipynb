```python
#Q1

import pandas as pd

file_path = 'kosaf_scholarship_2021.csv'
df = pd.read_csv(file_path, encoding='euc-kr')

numeric_cols = df.columns[5:]
df[numeric_cols] = df[numeric_cols].replace(',', '', regex=True)
df[numeric_cols] = df[numeric_cols].apply(pd.to_numeric, errors='coerce')

def money_format(x):
    return f"{int(x):,} 원"

name = input("학교명을 입력하세요. : ")
a_result = df[df['대학명'] == name].iloc[0]
print("Q1-(a)", name, "2021년 장학금 현황:")

외부장학금항목 = [
    ('국가', a_result['교외장학금 국가']),
    ('지방자치단체', a_result['교외장학금 지방자치단체']),
    ('사설 및 기타', a_result['교외장학금 사설및기타']),
]
print(f"1. 교외장학금 소계: {money_format(a_result['교외장학금 소계(원)'])}")
for i, (name, value) in enumerate(외부장학금항목, start=1):
    print(f"  {name}: {money_format(value)}")

내부장학금항목 = [
    ('성적우수', a_result['교내장학금 성적우수장학금']),
    ('저소득층', a_result['교내장학금 저소득층장학금']),
    ('근로', a_result['교내장학금 근로장학금']),
    ('기타', a_result['교내장학금 기타']),
]
print(f"2. 교내장학금 소계: {money_format(a_result['교내장학금 소계(원)'])}")
for i, (name, value) in enumerate(내부장학금항목, start=1):
    print(f"  {name}: {money_format(value)}")

print(f"3. 총계: {money_format(a_result['교외장학금 소계(원)'] + a_result['교내장학금 소계(원)'])}\n")

b_result = df['교외장학금 국가'].mean()
print(f"Q1-(b) 교외장학금 국가 평균: {money_format(b_result)}")

c_result = df[df['설립별'] == '사립']['교내장학금 성적우수장학금'].mean()
print(f"Q1-(c) 사립대 성적우수장학금 평균: {money_format(c_result)}")

private_avg = df[df['설립별'] == '사립']['교외장학금 소계(원)'].mean()
public_avg = df[df['설립별'] == '국공립']['교외장학금 소계(원)'].mean()
d_result = abs(private_avg - public_avg)
print(f"Q1-(d) 사립, 국공립 교외장학금 평균 차이: {money_format(d_result)}")

e_result = df[df['지역별'].isin(['대구', '경북'])]['교내장학금 근로장학금'].mean()
print(f"Q1-(e) 대구경북 근로장학금 평균: {money_format(e_result)}")
```

```python
#Q3

import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

plt.rc('font', family = 'NanumBarunGothic')
plt.rcParams['axes.unicode_minus'] =False

df_2021 = pd.read_csv("kosaf_scholarship_2021.csv", encoding='euc-kr')
df_2024 = pd.read_csv("kosaf_scholarship_2024.csv", encoding='euc-kr')

#(a)
knu_2021 = df_2021[df_2021['대학명'] == '경북대학교']
knu_2024 = df_2024[df_2024['학교명'] == '경북대학교']

categories = ['교외장학금 소계(원)', '교내장학금 소계(원)', '총계(원)']
values_2021 = knu_2021[categories].sum()
values_2024 = knu_2024[categories].sum()

x = range(len(categories))
width = 0.35

plt.figure(figsize=(10, 6))
bars_2021 = plt.bar([i - width/2 for i in x], values_2021, width=width, label='2021년', color='skyblue')
bars_2024 = plt.bar([i + width/2 for i in x], values_2024, width=width, label='2024년', color='lightgreen')

for bar in bars_2021:
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, height, f'{height:,.0f}', ha='center', va='bottom', fontsize=9)

for bar in bars_2024:
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, height, f'{height:,.0f}', ha='center', va='bottom', fontsize=9)

plt.xticks(x, categories)
plt.ylabel('장학금액 (원)')
plt.title('경북대학교 교외/교내/총 장학금 비교 (2021 vs 2024)')
plt.legend()
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()

#(b)~(e)
def compute_summary(df):
    result = {}
    result['(b)'] = df['교외장학금 국가'].mean()
    result['(c)'] = df[df['설립별'] == '사립']['교내장학금 성적우수장학금'].mean()
    private_avg = df[df['설립별'] == '사립']['교외장학금 소계(원)'].mean()
    public_avg = df[df['설립별'] == '국공립']['교외장학금 소계(원)'].mean()
    result['(d)'] = abs(private_avg - public_avg)
    result['(e)'] = df[df['지역별'].isin(['대구', '경북'])]['교내장학금 근로장학금'].mean()
    return result

result_2021 = compute_summary(df_2021)
result_2024 = compute_summary(df_2024)

def plot_comparison(result_2021, result_2024):
    labels = ['(b) 국가장학금 평균',
              '(c) 사립대 성적우수 평균',
              '(d) 사립/국공립 평균차이',
              '(e) 대구경북 근로장학금 평균']

    keys = ['(b)', '(c)', '(d)', '(e)']
    values_2021 = [result_2021[k] for k in keys]
    values_2024 = [result_2024[k] for k in keys]

    x = np.arange(len(labels))
    width = 0.35

    fig, ax = plt.subplots(figsize=(10, 6))
    bars1 = ax.bar(x - width/2, values_2021, width, label='2021년')
    bars2 = ax.bar(x + width/2, values_2024, width, label='2024년')

    ax.set_ylabel('장학금 평균액 (원)')
    ax.set_title('2021년 vs 2024년 장학금 항목별 비교')
    ax.set_xticks(x)
    ax.set_xticklabels(labels, rotation=15)
    ax.legend()

    def autolabel(bars):
        for bar in bars:
            height = bar.get_height()
            ax.annotate(f'{int(height):,}',
                        xy=(bar.get_x() + bar.get_width()/2, height),
                        xytext=(0, 3),
                        textcoords="offset points",
                        ha='center', va='bottom', fontsize=9)

    autolabel(bars1)
    autolabel(bars2)

    plt.tight_layout()
    plt.show()

plot_comparison(result_2021, result_2024)

print("(a) 경북대학교 장학금 현황 : 2024년에 전체 장학금이 증가, 특히 국가장학금과 근로장학금 항목 증가.")
print("(b) 전체 대학의 교외장학금(국가) 평균 : 2021년보다 2024년에 상승, 정부 장학금 지원 확대 추정.")
print("(c) 사립대학의 교내 성적우수장학금 평균 : 2024년에 상승, 학업 장려 목적의 재정 지원 강화 가능성.")
print("(d) 사립 vs 국공립 교외장학금 평균 차이 : 격차는 여전히 존재, 다만 2024년에 소폭 감소.")
print("(e) 대구·경북 지역 교내 근로장학금 평균 : 2024년에 증가, 지방대학 재정 지원 확대 흐름 반영.")
```

```python
import pandas as pd
import numpy as np
import pandas_datareader.data as web

df = web.DataReader('005930', 'naver', start='2024-01-01', end='2024-12-31')
df.head()
```

```python
#8 시작가('Open') 대비 종가('Close')가 높았던 날의 데이터

df['Open'] = pd.to_numeric(df['Open'], errors='coerce')
df['Close'] = pd.to_numeric(df['Close'], errors='coerce')

df[df['Open'] < df['Close']]
```

```python
#15 종가('Close') 데이터 가격이 높은 순서대로 정렬
df['Close'].sort_values(ascending=False)
```

```python
#16 인덱스 컬럼을 0부터 순차적으로 증가하는 정수로 변경
df.reset_index(inplace=True)
df
```

```python
#17 컬럼명을 Date는 '날짜'로, 'High'는 '고가'로, 'Low'는 '저가'로 변경
df.rename(columns={'Date': '날짜', 'High': '고가', 'Low': '저가'})
```

```python
#18 'Date' 칼럼 삭제
df.drop(columns=['Date'])
```

```python
#19 'diff' 컬럼 생성, 종가('Close')와 시가('Open')의 차이 계산하여 채우기
df['diff'] = df['Close']-df['Open']
df
```

```python
#20 'Date' 필드에 저장된 값을 파싱하여 'Year', 'Month', 'Day' 필드에 저장
df['Date'] = df['Date'].astype('datetime64[ns]')
df['Year'] = [d.year for d in df['Date']]
df['Month'] = [d.month for d in df['Date']]
df['Day'] = [d.day for d in df['Date']]
df
```

```python
#21 모든 데이터를 소수점 둘째자리까지 표시
numeric_cols = ['Open', 'High', 'Low', 'Close', 'Volume', 'diff']

for col in numeric_cols:
    if col in df.columns:
        df[col] = pd.to_numeric(df[col], errors='coerce')
        df[col] = df[col].apply(lambda x: "{:.2f}".format(x) if pd.notna(x) else x)

df
```

```python
#22 거래량 필드를 생성하고 세 자릿수마다 콤마 삽입
df['Volume'] = pd.to_numeric(df['Volume'], errors='coerce')
df['거래량'] = df['Volume'].apply(lambda x: "{:,}".format(x))
df
```

```python
#23 주가 데이터를 월별로 그룹핑, 거래량의 평균값 표시
volume_mean = df['Volume'].groupby(df['Month']).mean()
volume_mean
```

```python
#24 결측치가 있는지 확인, 있으면 처리
df.isnull().sum()
```

```python
#25 7월의 데이터와 8월의 데이터를 별도의 dec_df, nov_df으로 만들고 하나의 df로
dec_df = df[df['Month']==12]
nov_df = df[df['Month']==11]

nov_dec_df = pd.concat([nov_df, dec_df])
nov_dec_df
```
